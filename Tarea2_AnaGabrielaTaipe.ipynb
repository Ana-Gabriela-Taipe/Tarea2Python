## ¿Cómo calculan la medida de tendencia central y dispersión de la felicidad a nivel global para el primer y último año incluido en el set de datos?

vemos los valores únicos de los años presentes en el data set

```python
felicidad["anio"].unique
```

como vemos que el primer año es 2009 y el último año es 2018

```python
felicidad_anio = felicidad[(felicidad["anio"]==2009) | (felicidad["anio"]==2018)]
felicidad_anio
```

vemos como el data set de una dimensión de

```python
felicidad.shape
```

se redujo a 

```python
felicidad_anio.shape
```

Ahora, realicemos el Análisis Descriptivo de estos datos

```python
felicidad_anio.describe()
```

Recordemos que la felicidad o también conocida como Escalera de Vida indica la percepción de calidad de vida en una escala de 0 a 10.  Mientras más alto el puntaje, mayor felicidad. <br>
1. Como vemos el promedio de felicidad en nuestra muestra es de 5.48, es decir una felicidad media, ya que la escala es de 0 a 10. La desviación estandar es de 1.07 lo que significa que la mayoría de los puntajes de felicidad se encuentran dentro del intervalo $[\mu-2\sigma,\mu+2\sigma] = [3.34,7.62]$, lo cual indica que la mayoría de los paises de esta muestra tienen niveles de felicidad relativamente cercanos al promedio. <br>
2. Vemos también que el mínimo puntaje de felicidad es casi 3 puntos y el máximo es casi 8 puntos. Esto significa que hay variación en la percepción de calidad de vida entre los paises, teniendo algunos niveles bastantes bajos y otros muy altos.
3. Respecto a los rangos intercuartílicos que proporcionan información sobre la dispersión de los datos:
   - El 25% de la muestra tiene un valor de felicidad menor o igual a 4.68 puntos.
   - La mediana, que representa el 50% de la muestra, tiene un valor de felicidad menor o igual a 5.39 puntos, lo que indica que la mitad de los paises en nuestra muestra tienen un nivel de felicidad por debajo de este valor.
   - El 75% de la muestra tiene a lo más valor de 6.28 puntos.

En resumen, los paises en nuestra muestra tienen niveles de felicidad cercanos al promedio, pero existen diferencias significativas en paises, con algunos experimentando niveles de felicidad más bajos y otros más alto.

## ¿Cómo acceden a la información de la felicidad del primer y último año para los dos países de su elección?

Veamos que paises existen en la muestra filtrada por el primer y ultimo año

```python
felicidad_anio['pais'].unique()
```

eligiremos `El Salvador` y `Ecuador`

```python
felicidad_dos = felicidad_anio[(felicidad_anio['pais']=='Ecuador') | (felicidad_anio['pais']=='El Salvador')] 
felicidad_dos
```

```python
felicidad_dos.describe()
```

como vemos la media en términos de felicidad de estos paises está en 6.32, es decir, mayor al promedio 5.48 de felicidad a nivel mundial de la muestra `felicidad_anio` y ligeramente mayor al 75% de la muestra cuyo valor era de 6.28 puntos. Por tanto, estos dos paises a elección sí son más felices en estos años que el resto de paises en los mismos años, 2009 y 2018.
